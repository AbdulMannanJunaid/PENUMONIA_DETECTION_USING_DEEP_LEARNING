```python
from pydub import AudioSegment
import os

# Define the paths to your input and output directories
input_dir_normal = "/content/drive/MyDrive/Main/normal"
input_dir_pneumonia = "/content/drive/MyDrive/Main/pneumonia"
output_dir_normal = "/content/drive/MyDrive/Main/converted/normal"
output_dir_pneumonia = "/content/drive/MyDrive/Main/converted/pneumonia"

# Create the output directories if they don't exist
os.makedirs(output_dir_normal, exist_ok=True)
os.makedirs(output_dir_pneumonia, exist_ok=True)

# Function to convert audio files in a directory
def convert_audio_files(input_dir, output_dir):
    for filename in os.listdir(input_dir):
        if filename.endswith(".mp3"):
            # Load the MP3 file
            audio = AudioSegment.from_mp3(os.path.join(input_dir, filename))

            # Define the output file name with a WAV extension
            output_filename = os.path.splitext(filename)[0] + ".wav"

            # Save the audio as WAV in the output directory
            audio.export(os.path.join(output_dir, output_filename), format="wav")

# Convert and save normal audio files
convert_audio_files(input_dir_normal, output_dir_normal)

# Convert and save pneumonia audio files
convert_audio_files(input_dir_pneumonia, output_dir_pneumonia)
```
